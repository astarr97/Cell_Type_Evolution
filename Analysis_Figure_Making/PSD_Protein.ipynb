```python
import numpy as np
import math
import matplotlib.pyplot as plt
import csv
import pandas as pd
import re
from scipy.stats import pearsonr
#import rpy2.robjects as robjects
import random
from statsmodels.stats.multitest import fdrcorrection
import copy
from collections import Counter
import seaborn as sns
from scipy.stats import binom_test
import gseapy as gs
from scipy.stats import ttest_ind
from scipy.stats import spearmanr
```

```python
v = pd.read_csv("Human_Protein_Abundances.csv").set_index("Gene name")
vm = pd.read_csv("Macaque_Protein_Abundances.csv").set_index("Gene name")
vmo = pd.read_csv("Mouse_Protein_Abundances.csv").set_index("Gene name")
keep = []
for i in v.columns:
    if "Year18" in i:
        keep.append(i)
v_adult = v[keep].copy()
keep = []
for i in vm.columns:
    if "Year8" in i:
        keep.append(i)

vm_adult = vm[keep].copy()

keep = []
for i in vmo.columns:
    if "P36" in i:
        keep.append(i)

vmo_adult = vmo[keep].copy()
```

```python
gene = "DLG4"
gene_mou = gene[0] + gene[1:].lower()
mac = pd.DataFrame(vm_adult.loc[gene])
mac["Species"] = "Rhesus"
mac.columns = [gene, "Species"]
mac[gene] = 2**mac[gene]


hum = pd.DataFrame(v_adult.loc[gene])
hum["Species"] = "Human"
hum.columns = [gene, "Species"]
hum[gene] = 2**hum[gene]
mou = pd.DataFrame(vmo_adult.loc[gene_mou])
mou["Species"] = "Mouse"
mou.columns = [gene, "Species"]
mou[gene] = 2**mou[gene]
new_palette = {"Human":"#FF2C0C", "Chimp":"#0058FF", "Rhesus":"#AD6D3E", "Marmoset":"#551A7F", "Gorilla":"#5C687D", "Mouse":"#CAB2D6"}

sns.set(font_scale = 1.4)
sns.set_style("white")
fig, ax = plt.subplots(figsize=(7,5))

to_plot = pd.concat([hum, mac, mou])
#out.append([gene, np.mean(hum[gene]), np.mean(mac[gene]), np.mean(mou[gene]), ttest_ind(hum[gene], mac[gene])[1], ttest_ind(hum[gene], mou[gene])[1]])
sns.swarmplot(data = to_plot, y = gene, x = "Species", hue = "Species", palette = new_palette, size = 8)
dfp = pd.DataFrame([["Human", np.mean(hum[gene])], ["Rhesus", np.mean(mac[gene])], ["Mouse", np.mean(mou[gene])]])
dfp.columns = ["Species", "Abundance"]
t_ax = sns.barplot(data = dfp, x = "Species", y = "Abundance", hue = "Species", errorbar=None, linewidth=2.5, edgecolor=".5", facecolor='#F2C91140', gap = 0.3, palette = new_palette)
c = 0
for patch in t_ax.patches:
    if c < 1:
        patch.set_edgecolor("#FF2C0C")
        patch.set_facecolor("#FF2C0C1A")
    elif c == 1:
        patch.set_edgecolor("#AD6D3E")
        patch.set_facecolor("#AD6D3E1A")
    else:
        patch.set_edgecolor("#CAB2D6")
        patch.set_facecolor("#CAB2D61A")
    c += 1
plt.title("PSD-95 expression in PSD")
plt.ylabel("PSD-95 abundance in PSD")
```
